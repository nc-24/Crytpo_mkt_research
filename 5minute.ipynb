```python
from datetime import datetime, timedelta
import pandas as pd


def generate_range(start_date, end_date):
    delta = end_date - start_date
    return [start_date + timedelta(days=i) for i in range(delta.days + 1)]
def generate_kline(kline_dir,coin,start_date,end_date,interval_str):
    kline_file = kline_dir + "/BINANCEFUTURES_PERP_" + coin + "_USDT/" + coin + "USDT-"+interval_str+"-" + start_date.strftime("%Y-%m-%d") + ".csv"
    kline_df = pd.read_csv(kline_file)
    kline_df['ret'] = (kline_df['close'] - kline_df.shift(1)['close'])/kline_df.shift(1)['close']
    dates = generate_range(start_date,end_date)
    for date in dates[1:]:
        kline_file2 = kline_dir + "/BINANCEFUTURES_PERP_" + coin + "_USDT/" + coin + "USDT-"+interval_str+"-" + date.strftime("%Y-%m-%d") + ".csv"
        kline_df2 = pd.read_csv(kline_file2)
        kline_df2['ret'] = (kline_df2['close'] - kline_df2.shift(1)['close'])/kline_df2.shift(1)['close']
        kline_df2['ret'][0] = (kline_df2['close'][0] - kline_df['close'].iloc[-1])/kline_df['close'].iloc[-1]
        kline_df = pd.concat([kline_df,kline_df2])
    return kline_df
```

```python
import os as os
def generate_tweets_df(tweet_dir):
    tweets_file = os.path.join(tweet_dir, "tweets", "0.csv")
    tweets_df = pd.read_csv(tweets_file)
    tweets_df['acc_id']= 0
    file_list = sorted(os.listdir(os.path.join(tweet_dir, "tweets")))

    for filename in file_list:
        if filename.endswith(".csv") and filename != "0.csv":
            acc_id = filename.replace(".csv", "")
            file_path = os.path.join(tweet_dir, "tweets", filename)
            tweets_df_2 = pd.read_csv(file_path)
            tweets_df_2['acc_id'] = acc_id
            tweets_df = pd.concat([tweets_df, tweets_df_2])
    tweets_df['createdAt'] = pd.to_datetime(tweets_df['createdAt'], format="%a %b %d %H:%M:%S %z %Y")
    print(tweets_df)

    return tweets_df
```

```python
def calc_sharpe(tweets_df,ac):
    arr = {}
    for _, row in tweets_df.iterrows():
        acc_id = int(row['acc_id'])
        ret = row['ret']
        if not np.isnan(ret): 
            if acc_id not in arr:
                arr[acc_id] = []
            arr[acc_id].append(ret)
    
    means = {acc_id: np.mean(ret_list) for acc_id, ret_list in arr.items()}
    stddevs = {acc_id: np.std(ret_list) for acc_id, ret_list in arr.items()}
    sharpe_ratios = {acc_id: np.mean(ret_list) / np.std(ret_list) if np.std(ret_list) != 0 else np.nan
                     for acc_id, ret_list in arr.items()}
    ac_path = "/Users/nayanchauhan/Documents/Quant_Project/twitter/account.csv"
    ac = pd.read_csv(ac_path)
    
    ac.set_index('index',inplace=True)
    print("Mean:")
    for acc_id, mean in means.items():
        if acc_id in ac.index:
            account_name = ac.loc[acc_id, 'account']
            print(f"{account_name}: {mean}")
        else:
            print(f"error")
    print("\nStandard Deviation:")
    for acc_id, stddev in stddevs.items():
        if acc_id in ac.index:
            account_name = ac.loc[acc_id, 'account']
            print(f"{account_name}: {stddev}")
        else:
            print(f"error")
    print("\nSharpe Ratio:")
    for acc_id, sharpe in sharpe_ratios.items():
        if acc_id in ac.index:
            account_name = ac.loc[acc_id, 'account']
            print(f"{account_name}: {sharpe}")
        else:
            print(f"error")
```

```python
from transformers import pipeline
import pandas as pd
sentiment_pipeline = pipeline("sentiment-analysis")
pipe = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest",tokenizer="cardiffnlp/twitter-roberta-base-sentiment-latest", device =0)


def add_sentiment_features(df, sentiment_pipe):
    # Analyze sentiment for each text
    sentiments = df['text'].apply(lambda text: sentiment_pipe([text[:250]])[0])
    
    # Extract label and score, and adjust score if label is NEGATIVE
    df['label'] = sentiments.apply(lambda x: x['label'])
    df['sentiment_score'] = sentiments.apply(lambda x: x['score'])
    
    # Adjust sentiment_score to be negative if the label is NEGATIVE
    df['sentiment_score'] = df.apply(lambda row: -row['sentiment_score'] if row['label'] == 'NEGATIVE' else row['sentiment_score'], axis=1)
    
    return df
```

```python

add_sentiment_features(tweets_df, pipe)
```

```python
def add_sharpe(tweets_df, ac, col_index):
    index_to_name = {
        1: 'Positive_Sharpe_Ratio',
        2: 'Negative_Sharpe_Ratio',
        3: 'Neutral_Sharpe_Ratio'
    }
    
    arr = {}
    for _, row in tweets_df.iterrows():
        acc_id = int(row['acc_id'])
        ret = row['ret']
        if not np.isnan(ret): 
            if acc_id not in arr:
                arr[acc_id] = []
            arr[acc_id].append(ret)
    sharpe_ratios = {acc_id: np.mean(ret_list) / np.std(ret_list) if np.std(ret_list) != 0 else np.nan
                     for acc_id, ret_list in arr.items()}
    column_name = index_to_name.get(col_index, f'Sharpe_Ratio_{col_index}')
    if column_name not in ac.columns:
        ac[column_name] = np.nan
    for acc_id, sharpe_ratio in sharpe_ratios.items():
        if acc_id in ac.index:
            ac.at[acc_id, column_name] = sharpe_ratio
```
